```python
import numpy as np

data = np.load("/Users/maggiebowen/Documents/Dugue-Research/dyntex_python/demo/stimulus_fspatial_2.2.npz")

print(data.files)  # list of array names

for k in data.files:
    arr = data[k]
    print("\n---", k, "---")
    print("shape:", arr.shape)
    print("dtype:", arr.dtype)
```

```python
for k in data.files:
    arr = data[k]
    if np.issubdtype(arr.dtype, np.number):
        print(k, "min:", arr.min(), "max:", arr.max(), "mean:", arr.mean())
```

```python
for k in data.files:
    arr = data[k]
    
    if arr.dtype == object:
        print(k, "contains Python objects")
    
    if arr.ndim == 2:
        print(k, "likely tabular")
    
    if arr.ndim == 3:
        print(k, "likely image / sequence")
    
    if arr.ndim == 4:
        print(k, "likely batch of images / tensors")
```

```python
for k in data.files:
    arr = data[k]
    print(k, arr.flatten()[:10])
```

```python
for k in data.files:
    arr = data[k]
    if arr.ndim == 1 and len(np.unique(arr)) < 20:
        print(k, "likely labels:", np.unique(arr))
```

```python
import numpy as np

z = np.load("/Users/maggiebowen/Documents/Dugue-Research/dyntex_python/demo/stimulus_orientation_44.npz", allow_pickle=True)

feature_name = z["feature_name"]
profile = z["profile"]
fps = z["fps"]

print("feature:", feature_name)
print("fps:", fps)
print("start:", profile[0])
print("end:", profile[-1])

d = np.diff(profile)

if np.all(d == 0):
    print("direction: constant")
elif np.all(d[d != 0] > 0):
    print("direction: increasing")
elif np.all(d[d != 0] < 0):
    print("direction: decreasing")
else:
    print("direction: mixed")

change_idx = np.where(d != 0)[0]
if len(change_idx) > 0:
    print("first changing frame:", change_idx[0] + 1)
    print("last changing frame:", change_idx[-1] + 1)
else:
    print("no changing frames")
```
