# Runtime check

```python
import sys
import os
sys.path.append(os.path.abspath('..'))
from dyntex.MotionCloud import MotionCloud
import time

def to_numpy(x):
    return x.detach().cpu().numpy()
```

```python
for i in range(10):
    # Single MotionCloud
    N = 512
    frame_per_second = 100
    MC1 = MotionCloud(N=N, frame_per_second=frame_per_second,
                     contrast=35, ave_lum=127.0, 
                     over_samp=2, verbose=1)
    
    choose_kernel = 'bandpass'
    sf = 1.0
    sf_sig = 2.5
    th = 0.0
    th_sig = 10.0
    tf = 1.0
    spd_scalar = 1
    spd_dir = 0
    speed_sig = 50.0
    octa = 1
    
    MC1.set_all(choose_kernel, sf, sf_sig, th, th_sig, tf, spd_scalar, spd_dir, octa)
    MC1.burnout()
    
    print(MC1.dev)
    
    start = time.time()
    for i in range(10):
        MC1.get_frame(True)
    
    end = time.time()
    print(end-start)
```

```python
# Composite MotionCloud Check
for i in range(10):
    N = 256
    frame_per_second = 100
    percentage1 = 50
    percentage2 = 50
    
    MC1 = MotionCloud(N=N, frame_per_second=frame_per_second,
                     contrast=35, ave_lum=127.0, 
                     over_samp=2, verbose=1)
    
    MC2 = MotionCloud(N=N, frame_per_second=frame_per_second,
                     contrast=35, ave_lum=127.0, 
                     over_samp=2, verbose=1)
    
    choose_kernel = 'bandpass'
    sf = 1.0
    sf_sig = 2.5
    th = 0.0
    th_sig = 10.0
    tf = 1.0
    spd_scalar = 1
    spd_dir = 0
    speed_sig = 0.0
    octa = 1
    
    print(MC1.dev)
    MC1.set_all(choose_kernel, sf, sf_sig, th, th_sig, tf, spd_scalar, spd_dir, octa)
    MC1.burnout()
    
    choose_kernel = 'bandpass'
    sf = 1.0
    sf_sig = 2.5
    th = 0.0
    th_sig = 10.0
    tf = 1.0
    spd_scalar = 1
    spd_dir = 0
    speed_sig = 50.0
    octa = 1
    
    print(MC2.dev)
    MC2.set_all(choose_kernel, sf, sf_sig, th, th_sig, tf, spd_scalar, spd_dir, octa)
    MC2.burnout()
    
    start = time.time()
    for i in range(10):
        frame1 = MC1.get_frame(True)
        frame2 = MC1.get_frame(True)
        compositeFrame = (percentage1/100)*frame1 + (percentage2/100)*frame2
        MC1.update_composite_frame_hist(compositeFrame)
    end = time.time()
    print(end-start)
```
